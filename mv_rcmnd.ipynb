```python
import pandas as pd
import numpy as np
import ast
```

```python
df2 = pd.read_csv('tmdb_5000_movies.csv')
```

```python
cred = pd.read_csv('tmdb_5000_credits.csv')
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df.head()
```

```python
df = df.drop(columns = ['budget', 'homepage','original_language',
       'original_title','popularity', 'production_companies',
       'production_countries', 'release_date', 'revenue', 'runtime',
       'spoken_languages', 'status', 'tagline','vote_average',
       'vote_count'])
```

```python
df.head()
```

```python
def f1(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
df['genres'] = df['genres'].apply(f1)
```

```python
df.head()
```

```python
def f2(obj):
    L = []
    count = 0

    for i in ast.literal_eval(obj):
        if count < 3:
            L.append(i['name'])
        count+=1
    return L       
        
   


    
```

```python
df['keywords']=df2['keywords'].apply(f2)
```

```python
df.head()
```

```python
df = df.merge(cred, on = "title")
```

```python
df.head()
```

```python
def f3(obj):
    L = []
    

    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            L.append(i['name'])
        
    return L    
```

```python
df['cast'] = df['cast'].apply(f2)
```

```python
df.head()
```

```python
df['crew'] =df['crew'].apply(f3)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df['mrks'] = df['cast'] + df['crew'] + df['keywords'] + df['overview'] + df['genres']
```

```python
df= df.drop(columns="tag")
```

```python
df['mrks'][0]
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df['mrks'].apply(collapse)
```

```python
df['mrks'] = df['mrks'].apply(lambda x: " ".join(x))
```

```python
df =df.drop(columns=['crew', 'cast', 'genres', 'movie_id','keywords','overview'])
```

```python

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(df['mrks']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sm = cosine_similarity(vector)
```

```python
sm
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(sm[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:10]:
        print(df.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
